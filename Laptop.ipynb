```python
# Process Memory column
def extract_memory(memory_str, keyword):
    """Extract memory size in GB from the 'Memory' column."""
    memory_str = memory_str.upper()
    if keyword in memory_str:
        parts = memory_str.split()
        for i in range(len(parts)):
            if keyword in parts[i]:
                size = parts[i - 1]  # Get the preceding value
                size = size.replace('TB', '').replace('GB', '')
                try:
                    size = float(size)  # Convert to float first
                    if 'TB' in parts[i - 1]:
                        return int(size * 1024)  # Convert TB to GB
                    return int(size)
                except ValueError:
                    return 0
    return 0

df['SSD'] = df['Memory'].apply(lambda x: extract_memory(x, 'SSD'))
df['HDD'] = df['Memory'].apply(lambda x: extract_memory(x, 'HDD'))
df.drop(columns=['Memory'], inplace=True)
```

```python
X = df.drop(columns=['Price'])

y = np.log(df['Price'])

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

Random Forest Algorithm

```python
step1 = ColumnTransformer(transformers=[

('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,1,7,10,11])

],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,

random_state=3,

max_samples=0.5,

max_features=0.75,

max_depth=15)

pipe = Pipeline([

('step1',step1),

('step2',step2)

])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))

print('MAE',mean_absolute_error(y_test,y_pred))
```
